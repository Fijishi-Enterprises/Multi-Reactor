```python
from __future__ import division
import numpy as np
from scipy.integrate import solve_ivp
from scipy import interpolate 
import matplotlib.pyplot as plt
import pandas as pd
import datetime
from datetime import date
import datetime
from myfunctions_multi_scale import *
%matplotlib inline
```

```python
# Import of I data
# irradiance data is based on meteorological data from: https://ims.data.gov.il/he/ims/6 
dFI = pd.read_csv('../data/ims_data_2014_umol_photons.csv',encoding= 'unicode_escape',header=None)
days = list(range(1,366))
```

```python
# assign light data into an array for annual T and I plot

I_plot = np.zeros(24*365)
light_hours = list(range(5,19))
for day in days:
    for hour in light_hours:
        I_plot[(day - 1)*24 + hour] = float(dFI.iloc[day][hour-1])

annual_hours = list(range(1,(366-1)*24+1))
f0 = interpolate.interp1d(annual_hours, I_plot,kind = 'linear')
```

```python
# T and light data

x1 = pd.ExcelFile('../data/input.xlsx')

dfT = x1.parse('T_multi-scale',header=None)
# Temperatures are water temperatures, taken from:
# Y. Suari, et al., Sandbar Breaches Control of the Biogeochemistry of a Micro-Estuary RIME-restoration of
# Israeli micro estuaries. Front. Mar. Sci. (2019) 

days_reduced,hours_reduced = [],[]
for i in range(1,13):
    day = datetime.datetime(2014, i, 1)
    days_reduced.append(int(day.strftime("%j")))

days_reduced.append(366)
for j in range(1,13):
    day = datetime.datetime(2014, j, 1)
    hours_reduced.append((int(day.strftime("%j"))-1)*24)
hours_reduced.append((366-1)*24)
T = []

# assigmment of parameters to values: 
for key,val in zip(dfT.iloc[1:][0],dfT.iloc[1:][1]):
    exec(key + '=val')
    T.append(val)
    print(key,val)
T.append(T[0])
days = list(range(1,366))

f1 = interpolate.interp1d(hours_reduced, T,kind = 'cubic')

T_interp = f1(annual_hours)
print(len(T_interp))


fig, ax = plt.subplots(2,1,figsize=(16,10))
xlabels = ['1 Jan', '1 February', '1 March', '1 April', '1 May', '1 June', '1 July', '1 August', '1 September', '1 October','1 November', '1 December']

ax[0].plot(annual_hours,T_interp, 'r-')
ax[1].plot(annual_hours,I_plot, 'b-', linewidth=0.5)
ax[0].set_ylabel('Temperature \n[C]',fontsize=14, weight="bold")
ax[1].set_ylabel('Light Intensity \n[µmol photons $m^{-2}$ $second^{-1}$]',fontsize=14, weight="bold")
ax[0].set_xticklabels([])
ax[1].set_xticklabels([])
ax[1].set_xticks(annual_hours[0:len(annual_hours):int(len(annual_hours)/12-1)])
ax[1].set_xticklabels([str(i) for i in xlabels], rotation=45,fontsize=12, weight="bold")
```

```python
# assign light data into an array for simulations

I = np.zeros(24*365)
light_hours = list(range(5,19))
for day in days:
    for hour in light_hours:
        I[(day - 1)*24 + hour] = 0.43 * float(dFI.iloc[day][hour-1])

annual_hours = list(range(1,(366-1)*24+1))
f0 = interpolate.interp1d(annual_hours, I,kind = 'linear')
```

```python
x1 = pd.ExcelFile('../data/input.xlsx')
df1 = x1.parse('Parameters_multi-scale',header=None)

# assigmment of parameters to values: 
for key,val in zip(df1.iloc[:][0],df1.iloc[:][1]):
    exec(key + '=val')
    print(key,val)

n_reactors = np.int(n_reactors)
```

```python
# Reduction of Nsea to 10uM - 3 seasons (No summer)

Nsea_lim = 10 # uM N
n_reactors_lim = 750
NSEA, NEXT, NINT, M, Total_N, T = [],[],[],[],[],[]
Total_biomass = []
NSEA_seasons_f = []
Final_NSEA = []
NSEA_seasons, NINT_seasons, M_seasons = [],[],[]


print('Starting point')


# setup initial conditions identical for all reactors
x0 = n_reactors_lim*[Nsea, Next0, Nint0, m0]

dilution = 0
weeks = [1, 7, 45]
seasons = ['Winter', 'Spring', 'Autumn']
print(weeks)
for week0 in weeks:
    t0 = 10 + 7 * 24 * (week0-1)
    t = list(range(t0,t0 + int(n_days*24)))
    times = list(range(t0,t0 + int(n_days*24)))
    args = (Nintcrit,Nintmax,Nintmin,Vmax,Ks,KN,miu,S,Z,KI,K0,Ka,Topt,Tmin,Tmax,losses20,
            teta,Sopt,Smin,Smax, Qp, Qsea, Nsea,f1,f0,dilution,n,umol_to_percent_DW)

    # solve the ODEs using the new syntax
    sol = solve_ivp(multi_N_f_un, [t[0], t[-1]], x0, args=args, t_eval = t)

    # take the solution of the state variables:
    M_farm = []
    N_farm = []
    for i_reactor in range(n_reactors_lim):
        for i in range(i_reactor + 1):
            NSEA.append(sol.y[i*4,:])
            NEXT.append(sol.y[i*4+1,:])
            NINT.append(sol.y[i*4+2,:])
            M.append(sol.y[i*4+3,:])
            M_farm.append((M[-1][-1] - m0) * 1.785)
            N_farm.append((M[-1][-1] * NINT[-1][-1] - m0 * Nint0) * 1.785 / 100)
        Total_biomass.append(round(sum(M_farm),1))
        Total_N.append(round(sum(N_farm),2))

        T.append(sol.t)
        
        Final_NSEA.append(NSEA[-1][-1])
        if NSEA[-1][-1] < Nsea_lim:
            NSEA_seasons_f.append(Final_NSEA)
            Final_NSEA = []
            NSEA_seasons.append(NSEA[-1])
            NINT_seasons.append(NINT[-1])
            M_seasons.append(M[-1])
            
            plot_result_Nsea(T,NSEA,NEXT,NINT,M,Nsea,i_reactor)
            break
        NSEA, NEXT, NINT, M, N_farm, Total_N, T = [],[],[],[],[],[],[]
        
        x0 = n_reactors_lim*[Nsea, Next0, Nint0, m0]

print("Done")

lengths = []
```

```python
# define function - plotting NSEA, NINT and M vs time - cage N with Nsea < Nsea_lim in winter, for 3 seasons

def plot_result_seasons(T,NSEA_seasons,NINT_seasons,M_seasons,Nsea,NSEA_seasons_f,lengths):
    """ Plot time series of the results, it's organized as a list of solution structures, sol.t,sol.y 
    
    We will plot 3 axes each for the separate quantity, only the last reactors get a line
    """
    
    import matplotlib.pyplot as plt
    import numpy as np
    
    
    fig = plt.figure()
    fig,ax = plt.subplots(2,2,figsize = (16,9))
    plt.rcParams.update({'font.size': 11.5})

    xlabels = ['day 0', '', 'day 2', '', 'day 4', '', 'day 6', '', 'day 8', '','day 10', '','day 12','', 'day 14']
    t = T[0]
    I = []
    colors = ['blue', 'red', 'orange']
    seasons = ['Winter', 'Spring', 'Autumn'] 
    for i in seasons:
        ax.flat[2].plot(t,NSEA_seasons[seasons.index(i)],'-',color=colors[seasons.index(i)],markersize=2)
        ax.flat[2].set_ylim([0,Nsea*1.2])
        ax.flat[1].plot(t,NINT_seasons[seasons.index(i)],'-',color=colors[seasons.index(i)], markersize=2)
        ax.flat[1].set_ylim([1.5,4])
        ax.flat[3].plot(t,M_seasons[seasons.index(i)],'-',color=colors[seasons.index(i)], markersize=2)
        I.append(i)
        
    ax.flat[2].set_ylabel('Nenv \n [µmol $l^{-1}$]',fontsize=14, weight="bold")
    ax.flat[1].set_ylabel('Nint \n [% g N $g DW^{-1}$]',fontsize=14, weight="bold")
    ax.flat[3].set_ylabel('m \n [g DW $l^{-1}$]',fontsize=14, weight="bold")
    
    ax.flat[2].set_xticklabels([])
    ax.flat[3].set_xticklabels([])
    ax.flat[1].set_xticklabels([])
    ax.flat[2].set_xticks(np.linspace(T[0][0],T[0][0]+14*24,15))
    ax.flat[3].set_xticks(np.linspace(T[0][0],T[0][0]+14*24,15))
    ax.flat[1].set_xticks(np.linspace(T[0][0],T[0][0]+14*24,15))
    ax.flat[2].set_xticklabels([str(i) for i in xlabels], rotation=45,fontsize=12)
    ax.flat[3].set_xticklabels([str(i) for i in xlabels], rotation=45,fontsize=12)
    ax.flat[1].set_xticklabels([str(i) for i in xlabels], rotation=45,fontsize=12)
    ax.flat[0].set_xlabel('Number of reactors in farm',fontsize=14, weight="bold")
    ax.flat[0].set_ylabel('Final Nenv \n [µmol N $l^{-1}$]',fontsize=14, weight="bold")
    colors = ['blue', 'red', 'orange']
    seasons = ['Winter', 'Spring', 'Autumn']
    for i in range (3):
        x = range(len(NSEA_seasons_f[i]))
        print(len(NSEA_seasons_f[i]))
        lengths.append(len(x))
        y = NSEA_seasons_f[i]
        ax.flat[0].plot(x,y, '--',color=colors[i],markersize=1.5)
        ax.flat[0].set_xlim([0,1.1*max(lengths)])
    ax.flat[0].legend(seasons,loc='upper right',fontsize=14)
    name = 'fig5.png' 
    fig.savefig(fname=name, dpi=600)
    
    return lengths
```

```python
# N and biomass Dynamics in 3 seasons - Winter n_reactor (different cultivation lengths - 14,9,11)

n_reactors = max(lengths)
NSEA, NEXT, NINT, M, Total_N, T = [],[],[],[],[],[]
Total_biomass = []
NSEA_seasons_f = []
Final_NSEA = []
NSEA_seasons, NINT_seasons, M_seasons = [],[],[]
Data = []
n_days_all = [14, 9, 11]
T_all = []

print('Starting point')


# setup initial conditions identical for all reactors
x0 = n_reactors*[Nsea, Next0, Nint0, m0]

dilution = 0
weeks = [1, 7, 45]
seasons = ['Winter', 'Spring', 'Autumn']
print(weeks)

for week0 in weeks:
    n_days = n_days_all[weeks.index(week0)]
    t0 = 10 + 7 * 24 * (week0-1)
    t = list(range(t0,t0 + int(n_days*24)))
    times = list(range(t0,t0 + int(n_days*24)))
    args = (Nintcrit,Nintmax,Nintmin,Vmax,Ks,KN,miu,S,Z,KI,K0,Ka,Topt,Tmin,Tmax,losses20,
            teta,Sopt,Smin,Smax, Qp, Qsea, Nsea,f1,f0,dilution,n,umol_to_percent_DW)

    # solve the ODEs using the new syntax
    sol = solve_ivp(multi_N_f_un, [t[0], t[-1]], x0, args=args, t_eval = t)

    # take the solution of the state variables:

    for i_reactor in range(n_reactors):
        NSEA, NEXT, NINT, M, N_farm, Total_N, T, N_farm, M_farm = [],[],[],[],[],[],[],[],[]
        for i in range(i_reactor+1):
            NSEA.append(sol.y[i*4,:]) #[umol N in sea / l]
            NEXT.append(sol.y[i*4+1,:]) #[umol N in reactor / l]
            NINT.append(sol.y[i*4+2,:]) #[% g N / g DW]
            M.append(sol.y[i*4+3,:]) #[g DW / l]
            M_farm.append((M[-1][-1] - m0) * 1.785) #[produced kg DW / reactor / 2 weeks]
            N_farm.append((M[-1][-1] * NINT[-1][-1] - m0 * Nint0) * 1.785 / 100) #[removed g N / reactor eq sea volume / 2 weeks]
        Total_biomass.append(round(sum(M_farm),1)) #[produced kg DW / farm / 2 weeks]
        Total_N.append(round(sum(N_farm),2)) #[removed g N / farm / 2 weeks]
        T.append(sol.t)
    Seasonal_production = sum(Total_biomass)
    Seasonal_sequestration = sum(Total_N) #[removed g N / farm / 2 weeks]
    Data.append([Seasonal_production,Seasonal_sequestration])    
    NSEA_seasons.append(NSEA[-1])
    NINT_seasons.append(NINT[-1])
    M_seasons.append(M[-1])
    T_all.append(T)
    x0 = n_reactors*[Nsea, Next0, Nint0, m0]

T1 = T_all[0]
T2 = T_all[1]
T3 = T_all[2]
plot_result_seasons_days(T1,T2,T3,NSEA_seasons,NINT_seasons,M_seasons,Nsea)
np.savetxt('Seasonal_total_values_140911.csv',Data,delimiter=',')         

print("Done")

n_days = 14
```
